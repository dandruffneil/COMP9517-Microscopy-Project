```python
img_test = img_gray.copy()
bg_pixel = np.bincount(img_test.ravel()).argmax()
print(f'Background pixel value is {bg_pixel}')

_, img_th = cv2.threshold(img_gray, bg_pixel+1, 255, 0)
plt.imshow(img_th, cmap='gray'), plt.title(f'Thresholded at {bg_pixel+1}')
plt.show()

_, img_th = cv2.threshold(img_gray, bg_pixel, 255, 0)
plt.imshow(img_th, cmap='gray'), plt.title(f'Thresholded at {bg_pixel}')
plt.show()

_, img_th = cv2.threshold(img_gray, bg_pixel-1, 255, 0)
plt.imshow(img_th, cmap='gray'), plt.title(f'Thresholded at {bg_pixel-1}')
plt.show()
```

```python
def max_filter(N, I, O):
    # Zero-padding is used since the filter uses the maximum value
    height = len(I)
    width = len(I[0])
    rel_limit = int(N / 2)

    for i in range(height):
        for j in range(width):
            max_pixel = 0
            for a in range(-rel_limit,rel_limit+1):
                if i+a >= 0 and i+a < height:
                    for b in range(-rel_limit,rel_limit+1):
                        if j+b >= 0 and j+b < width:
                            if I[i+a][j+b] > max_pixel:
                                max_pixel = I[i+a][j+b]
            O[i][j] = max_pixel
```

```python
def min_filter(N, I, O):
    # Max-padding is used since the filter uses the minimum value
    height = len(I)
    width = len(I[0])
    rel_limit = int(N / 2)

    for i in range(height):
        for j in range(width):
            min_pixel = 255
            for a in range(-rel_limit,rel_limit+1):
                if i+a >= 0 and i+a < height:
                    for b in range(-rel_limit,rel_limit+1):
                        if j+b >= 0 and j+b < width:
                            if I[i+a][j+b] < min_pixel:
                                min_pixel = I[i+a][j+b]
            O[i][j] = min_pixel
```

```python
def remove_shading(M, N, I):
    A = I.copy()
    B = I.copy()

    # Task 1 - Background Estimation
    if M == 0:
        max_filter(N,I,A)
        min_filter(N,A,B)
        plt.imshow(B, cmap='gray'),plt.title('Min-filtered Image')
        plt.show()
    elif M == 1:
        min_filter(N,I,A)
        max_filter(N,A,B)
        plt.imshow(B, cmap='gray'),plt.title('Max-filtered Image')
        plt.show()

    # Task 2 - Background Subtraction
    O = I.copy()

    for i in range(len(I)):
        for j in range(len(I[0])):
            i_pixel = int(I[i][j])
            b_pixel = int(B[i][j])
            if b_pixel >= i_pixel:
                if M == 0:
                    O[i][j] = i_pixel - b_pixel + 255
                elif M == 1:
                    O[i][j] = i_pixel - b_pixel
            else:
                O[i][j] = i_pixel - b_pixel

    return O
```

```python
# Algorithm
M = 0
N = 40
img_max_filter = remove_shading(M,N,img_gray.copy())
plt.imshow(img_max_filter, cmap='gray'),plt.title('Output Max-Min Filter')
plt.show()

# For cells:
M = 1
N = 40
img_min_filter = remove_shading(M,N,img_gray.copy())
plt.imshow(img_min_filter, cmap='gray'),plt.title('Output Min-Max Filter')
plt.show()

```
